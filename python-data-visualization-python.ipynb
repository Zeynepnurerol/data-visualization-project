```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 
df = pd.read_csv(r"C:\Users\GOKHAN\Desktop\2015.csv")
print(df.head(7))

#ÇİZGİ GRAFİĞİ
ulke=df["Country"].head(7)
mutluluk_skoru=df["Happiness Score"].head(7)
plt.plot(ulke,mutluluk_skoru,marker="o",color="green",linewidth=4)
plt.title("Ülkelere Göre Mutluluk Skorları")
plt.xlabel("Ülkeler")
plt.ylabel("Mutluluk Skorları")
plt.grid(True)
plt.xticks(rotation=45, ha='right')
plt.show()


#Grafiğin Yorumu:
#Switzerland en yüksek mutluluk skoruna sahip ülkedir.
#Hollanda en düşük mutluluk skoruna sahip ülkedir.
#Soldan sağa doğru azalan mutluluk skoru görünüyor.
#Norway'den Canada'ya geçerken diğerlerine göre daha sert bir düşüş vardır.
#İlk üç ülke mutlulukts öne çıkmaktadır.
```

```python
#DAĞILIM GRAFİĞİ

bolge= df["Region"].head(15)
kişi_başı_milli_ekonomi= df["Economy (GDP per Capita)"].head(15)
plt.scatter(bolge,kişi_başı_milli_ekonomi, color='blue', marker='o')
plt.title("Bölgelere Göre Kişi Başı Milli Ekonomi Dağılımı")
plt.xlabel("Bölge")
plt.ylabel("Ekonomi")
plt.grid(True)
plt.xticks(rotation=45, ha='right')
plt.show()


#Grafiğin Yorumu:
#Western Europe'da noktaların kümelenmiş olma sebebi birçok kişinin kişi başı ekonomik skorlının çok yakın olmasıdır.
#Western Europe diğer bölgelere göre ekonomik gelişmişlik seviyesi yüksektir.
#Latin America and Carriben bölgesi en düşük ekonomik seviyeye sahip bölgedir.
#Latin America and Carriben bölgesinde değerlerin geniş aralığa yayılması bçlgesel ekonomik eşitsizliği göstermektedir.
```

```python
#BAR GRAFİĞİ

ulke=df["Country"].head(9)
saglik_yasam_beklentisi = df["Health (Life Expectancy)"].head(9)
plt.figure(figsize=(8, 6))
plt.bar(ulke, saglik_yasam_beklentisi, color="cornflowerblue")
plt.title("Ülkelere Göre Sağlık")
plt.xlabel("Ülke")
plt.ylabel("Sağlık")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.xticks(rotation=50, ha='right')
plt.show()

#Grafiğin Yorumu:
#Grfikte gözüken tüm ülkelerin sağlık skorları yüksek seviyededir.
#En yüksek sağlık skoruna sahip ülke Icelan.
#Skorların bu kadar yakın olma sebebi hepsininn gelişmiş ülke olması ve sağlık standartlarının benzemesidir.
#En düşük skor bile seçilen ülkeler açısından homojen olduğunu gösteriyor.
```

```python
#KORELASYON ISI HARİTASI
veri = {
    "Saglık": df["Health (Life Expectancy)"].head(9),
    "Mutluluk_Sıralaması":df["Happiness Rank"].head(9),
    "Ozgurluk":df["Freedom"].head(9),
    "Guven":df["Trust (Government Corruption)"].head(9),
    "Comertlik":df["Generosity"].head(9),
}

df=pd.DataFrame(veri)
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Korelasyon Grafiği")
plt.show()

#Grafiğin Yorumu:
#1'e yakın olması güçlü bir pozitif ilişki olduğunu göstermektedir.
#-1'e yakın olması negatif bir ilişki olduğunu göstermektedir.
#Sağlık değeri yüksek olan bölgelerde güven seviyesi daha düşüktür.
#Özgürlüğün olduğu yerde güvenin daha yüksek olduğu görülmektedir.
#Sağlık ile özgürlük arasında nerdeyse hiç ilişki yoktur.
#En güçlü negatif ilişki özgürlük ile cömertlik arasındadır.
```

```python
#PASTA GRAFİĞİ
ulke=df["Country"].head(7) 
ozgurluk = df["Freedom"].head(7)
plt.pie(
    ozgurluk,  
    labels=ulke,                             
    colors=["skyblue", "lightgreen", "lightcoral", "gold", "orange", "violet", "cyan"],
    autopct="%1.1f%%",
    startangle=90,
    shadow=True
)
plt.title("Ülkelerin Özgürlük Dağılımı")
plt.axis("equal")
plt.show()

#Grafiğin Yorumu:
#Ülkelerin özgürlük oranları birbirine yakındır.
#Özgürlük payı en yüksek ülke Norveç'tir.
#En düşük pay Netherlands'a aittir.
#Seçilen ülkeler özgürlük açısından Dünya'nın en yüksek puanlı ülkeleri olduğu için yüzdeler birbirine çok yakındır.
```

```python
#HİSTOGRAM GRAFİĞİ
guven = df["Trust (Government Corruption)"].head(30)
plt.hist(guven, bins=5, edgecolor="black")
plt.title("Ülkelerin Güven Dağılımı (Histogram Grafiği)")
plt.xlabel("Güven")
plt.ylabel("Ülke Sayısı") 
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

#Grafiğin Yorumu:
#En yoğun güven aralığında yaklaşık 8 ülke bulunmaktadır.
#Güven düzeyi arttıkça ülkelerin sayısı azalmaya başlamıştır.
#Dağılım sağa çarpık bir yapı gösteriyor.
#Veri seti yüksek güvene sahip ülklerin az, düşük güvenli ülkelerin ise çok olduğunu gösteriyor.
#Güven düzeyi düşük olan ülkeler çoğunluktadır.
```

```python
harita = folium.Map(location=[39.9334, 32.8597], zoom_start=6) 
folium.Marker(
    location=[41.3977409, 33.7057676],  
    popup="Kastamonu Şube",
    tooltip="Kastamonu",
    icon=folium.Icon(color="red",icon="heart")
).add_to(harita)
folium.Marker(
    location=[40.7328725, 29.811114],  
    popup="Sakarya Şube",
    tooltip="Sakarya",
    icon=folium.Icon(color="blue",icon="star")
).add_to(harita)
folium.Marker(
    location=[36.9975753, 35.1232531],  
    popup="Adana Şube",
    tooltip="Adana",
    icon=folium.Icon(color="green", icon="ok-sign")
).add_to(harita)
harita
```

```python
harita = folium.Map(location=[39.9334, 32.8597], zoom_start=6)
rota=[
 [40.7328725, 29.811114],
 [41.3977409, 33.7057676],
 [36.9975753, 35.1232531]
]
folium.PolyLine(rota, color="blue", weight=4, opacity=0.7).add_to(harita)
for sehir, konum in zip(["Sakarya", "Kastamonu", "Adana"], rota):  
    folium.Marker(location=konum, popup=sehir).add_to(harita)
harita
```
